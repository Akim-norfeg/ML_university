```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Функция одной пеерменной

```python
# Функция и её градиент
def f(x):
    return 3 * x**2

def df(x):
    return 6 * x

x0 = 5  #начальная точка
learning_rate = 0.1
tolerance = 1e-8
history = [(x0, f(x0))]
```

### Градиентный спуск

```python
while True:
    x1 = x0 - learning_rate * df(x0)
    history.append((x1, f(x1)))
    
    if np.abs(x1 - x0) < tolerance:
        break
    
    x0 = x1
```

### График сходимости

```python
history = np.array(history)
plt.plot(history[:, 0], history[:, 1], 'ro-', label="Путь сходимости")
x_vals = np.linspace(-5, 5, 100)
plt.plot(x_vals, f(x_vals), 'b-', label="Функция f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.title("Градиентный спуск для f(x) = 3x²")
plt.show()
```

## Функция двух переменных

### Функция и градиент

```python
def f2(x, y):
    return 3*x**2 + 5*y**2

def df2(x, y):
    return np.array([6*x, 10*y])

x0 = np.array([5, 10])  #начальная точка
learning_rate = 0.1
tolerance = 1e-8
history = [(*x0, f2(*x0))]
```

### Градиентный спуск

```python
while True:
    x1 = x0 - learning_rate * df2(*x0)
    history.append((*x1, f2(*x1)))
    
    if np.linalg.norm(x1 - x0) < tolerance:
        break
    
    x0 = x1
```

### График сходимости

```python
history = np.array(history)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

#поверхность функции
X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = f2(X, Y)
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

#путь градиентного спуска
ax.plot(history[:, 0], history[:, 1], history[:, 2], 'ro-', label="Путь сходимости")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.set_title("Градиентный спуск для f(x, y) = 3x² + 5y²")
plt.legend()
plt.show()
```
